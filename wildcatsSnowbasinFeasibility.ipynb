Part 1: The Problem & Impact
The “So What?” Question

If a 95% accurate model can predict peak-hour traffic volume on Trappers Loop (SR-167) 72 hours in advance, Snowbasin Resort operations managers can proactively adjust staffing levels, activate overflow parking, and communicate arrival recommendations to guests. This reduces congestion, improves guest experience, and lowers operational risk during high-demand days.

The Target Variable

The model predicts peak-hour traffic volume (continuous numeric value), defined as the number of vehicles passing the Trappers Loop traffic station during the morning arrival window (7:00–9:00 AM). A secondary classification output may flag “High Congestion” vs “Normal” days as a binary variable for operational decision-making.

Part 2: The Data Audit (Verification of Life)
Dataset A (Primary Source)

Name / Origin: Utah Department of Transportation (UDOT) Traffic Statistics
Format: CSV
Description: Hourly vehicle counts from Traffic Station #029-0010 on Trappers Loop (SR-167)
Row Count: ~8,000–9,000 rows per year (24 hours × 365 days)
Access Status: Publicly available; sample CSVs downloadable from UDOT traffic statistics portal

Dataset B (Enrichment Source)

Name / Origin: NOAA Local Climatological Data (Ogden / Huntsville stations)
Format: CSV or JSON API
Description: Historical weather observations and short-term forecasts including snowfall, temperature, wind speed, and visibility
Row Count: Tens of thousands of hourly records across multiple years
Access Status: Publicly available; no API key required; sample data already accessible

Optional Dataset C (Categorical Enrichment)

Name / Origin: Utah State Holidays & Weber State University Academic Calendar
Format: Manually curated CSV
Row Count: <100 rows
Access Status: Publicly available calendars

Part 3: The Integration Strategy (The “Join”)
The Join Key

Date + Hour (Timestamp)

Granularity Check

All datasets are aggregated at an hourly time-series level, allowing for a direct and clean join without spatial transformations. Traffic volume, weather conditions, and holiday indicators align naturally on timestamp.

Part 4: Risk & The “Kill Switch”
The Technical Blocker

The primary technical challenge is aligning historical traffic data with weather forecasts at a consistent hourly resolution, particularly when handling missing values or station outages in NOAA data.

The Backup Plan

If traffic sensor data becomes incomplete or unreliable, the project will pivot to a binary congestion classifier that predicts “High” vs “Normal” traffic days using only weather conditions and calendar effects. This alternative still delivers clear operational value while reducing dependency on precise traffic counts.
